```python
#conversion PSI to Pa
psi_to_pa = 6894.76
M3_to_L = 1000
R = 8.3145
T = 300
Pressure = np.array([5e3*psi_to_pa,10e3*psi_to_pa,15e3*psi_to_pa]).reshape((-1, 1))
log_k_k0 = np.array([0.10,0.21,0.31])

plt.plot(Pressure,log_k_k0)
plt.ylabel("Log(k/K0)")
plt.xlabel("Pressure (Pa)")
```

```python
model = LinearRegression().fit(Pressure,log_k_k0)
print('intercept:', model.intercept_)
print('slope:', model.coef_)
```

```python
slope = model.coef_
Answer = (slope*(R*T))*M3_to_L
print("Product of formal charges:",Answer,"L/mol")
```
